```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#given variables
x_distance = 4
distance_traveled = 100
velocity = 10
```

```python
#use pythagorean therom to get distance from ambulance
N = 0.1 #sample rate
t = np.arange(0, 10 + N, N)
ambulance_range = ((4**2) + (50 - (t * velocity))**2)**(1/2)
plt.plot(t, ambulance_range)
plt.xlabel("Time(sec)")
plt.ylabel("Range")
plt.title("Range vs. Time Graph")
plt.show()
```

```python
radar_freq = 2.2 * 10**9
light_speed = 3 * 10**8
wavelength = light_speed/radar_freq
t = np.arange(0, 10 + N, N)

radial_vel = np.diff(ambulance_range)/-np.diff(t) #derivative bonus point

dopper_shift = -(2*radial_vel)/wavelength
plt.xlabel("Time(sec)")
plt.ylabel("Doppler Shift")
plt.title("Doppler Shift vs. Time Graph")
plt.plot(t[:-1], dopper_shift)
```

```python
#Acceleration bonus
velocity = np.linspace(10, 30, 101)
N = 0.1 #sample rate
t = np.arange(0, 10 + N, N)
ambulance_range = ((4**2) + (50 - (t * velocity))**2)**(1/2)
plt.xlabel("Time(sec)")
plt.ylabel("Range")
plt.title("Range vs. Time Graph (With Acceleration)")
plt.plot(t, ambulance_range)
```

```python
your_radar = 2.2 * 10**9
friend_radar = 10 * 10**9

your_wavelength = light_speed/your_radar
friend_wavelength = light_speed/friend_radar
```

```python
#escape velocities (m/s)
planets = ["mercury", 'venus', 'earth', 'moon', 'mars', 'jupiter', 'saturn', 'uranus', 'neptune']
escape_v = np.array([4.3, 10.4, 11.2, 2.4, 5.0, 59.5, 35.5, 21.3, 23.5]) * 10**3

your_shift = -(2 * escape_v)/your_wavelength
friend_shift = -(2 * (escape_v * np.cos(30)) )/friend_wavelength
```

```python
plt.xlabel("Planets")
plt.ylabel("Doppler Shift")
plt.title("Your Doppler Shift vs Planet")
plt.bar(planets, your_shift)
```

```python
plt.xlabel("Planets")
plt.ylabel("Doppler Shift")
plt.title("Friend Doppler Shift vs Planet")
plt.bar(planets, friend_shift)
```

# Applications of Doppler Effect

1.) Police Radars use the doppler effect to detect the speed of passing vehicles. When the radar sends waves towards the passing car, by the doppler effect the waves sent back will have a different frequency. A calculation is then done on these two frequencies to determine the speed of the passing car.

2.) When an ambulance drives by, most people can tell how close the ambulance is just from hearing the siren. This is because when the ambulance gets closer, by the doppler effect the frequency increases, which determines the pitch.

3.) The doppler effect is used for astronomy, specifically for determing the speed of a moving star in space.

4.) The doppler effect is used in the medical field through a test known as an echocardiogram. Doctors use this to determine the velocity of cardiac tissues and blood in one's body.

5.) The military uses the doppler effect to determine the speed of submarines and flying objects such as drones or planes.

```python
#6
N = 0.2 #sample rate
t = np.arange(0, 2, N)
Fc = 1
x = np.sin(2*np.pi*Fc * t)
plt.title('Sinusoidal Waves')
plt.xlabel("Time")
plt.ylabel('Amplitude')
plt.plot(t, x)
plt.show()
```

```python
#7
N = 0.2 #sample rate
t = np.arange(0, 2, N)
Fc = 1
x = np.cos(2*np.pi*Fc * t)
plt.scatter(t, x, c='blue', marker= "s")
plt.show()
#Difference is cosine starts at 1, sine starts at 0
```

```python
#8
N = 0.02 #sample rate
t = np.arange(8, 12, N)
Fc_cos = 1/1.5
Fc_sin = 1/0.5
x0 = 4 * np.cos(2*np.pi*Fc_cos * t)
x1 = 2 * np.sin(2*np.pi*Fc_sin * t)
y = x0 + x1
plt.title('Sum of Sine and Cosine Waves')
plt.xlabel("Time")
plt.ylabel('Amplitude')
plt.plot(t, y)
plt.show()
#2 and 2/3 periods are shown
```
